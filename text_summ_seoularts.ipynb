```python
# start code for summarizer from here

class Sentence:

    @staticmethod
    def co_occurence(sentence1, sentence2):
        p = sum((sentence1.bow & sentence2.bow).values())
        q = sum((sentence1.bow | sentence2.bow).values())
        return p / q if q else 0

    def __init__(self, data):
        self.index = data[0]
        self.text = data[1]
        self.token = data[2]
        self.bow = Counter(self.token)

    def __eq__(self, another):
        return hasattr(another, 'index') and self.index == another.index

    def __hash__(self):
        return self.index
    
def build_graph(sentences):
    graph = networkx.Graph()
    graph.add_nodes_from(sentences)
    pairs = list(itertools.combinations(sentences, 2))
    for eins, zwei in pairs:
        graph.add_edge(eins, zwei, weight=Sentence.co_occurence(eins, zwei))
    return graph

def summary(reviews):
    sentences = []
    for review in reviews:
        if review[1].strip():
            sentences.append(Sentence(review))
    with warnings.catch_warnings():
        warnings.simplefilter('ignore')
        graph = build_graph(sentences)
        pagerank = networkx.pagerank(graph, weight='weight')
        add_score = [(sentence, score) for sentence, score in pagerank.items()]
        reordered = sorted(add_score, key=lambda x: x[1], reverse=True)
#         reordered = sorted(pagerank, key=pagerank.get, reverse=True)

    return [(x.text, y) for x, y in reordered]
```

```python
from konlpy.tag import Mecab
from collections import Counter


def preprocessing(docs):
    mecab = Mecab()
    data = []
    for idx, doc in enumerate(docs):
        item = (idx, doc, mecab.nouns(doc))
        data.append(item)
    return data
```

```python
import itertools
import warnings
from collections import Counter
from collections import defaultdict

import networkx
```

```python
seoularts_by_date = defaultdict(list)
for i in data:
    seoularts_by_date[i['date'][:7]].append({'title':i['title'], 'body': i['body']})
```

```python
qna_cnt_by_date = []
for date, items in seoularts_by_date.items():
    qna_cnt_by_date.append((date, len(items)))
qna_reordered = sorted(qna_cnt_by_date, key=lambda x: int(str(x[0]).replace('-', '')))
```

# 2006년 8월 문의 전체 (43개 문의)

```python
def data_clean(item):
    return item['body'].replace('\t', '').replace('\n', '').replace('\xa0', '').replace('\r', '')

for i in [data_clean(b) for b in seoularts_by_date['2006-08']]:
    print(i)
    print('-'*40)
    print()
```

## 2006년 8월 문의 top5 (스코어 같이)

```python
i = 0
for text, score in summary(preprocessing([data_clean(b) for b in seoularts_by_date['2006-08']]))[:5]:
    if i == 0:
        sample = text
    print('문의내용\n', text)
    print('스코어:', score)
    print('-'*80)
    print()
    i += 1
```

## 2006년 8월 문의 중 키워드 랭크

```python
mecab = Mecab()
nouns = mecab.nouns(' '.join([data_clean(b) for b in seoularts_by_date['2006-08']]))
keywords_cnt = Counter([n for n in nouns if len(n)>1]).most_common(50)
keywords_cnt
```

## 2006년 8월 top1 문의 키워드 분석

```python
print(sample)
nouns = mecab.nouns(sample)
print()
print('-'*80)
print('사용된 키워드의 전체 출현 횟수')
for word, _ in Counter([n for n in nouns if len(n)>1]).most_common():
    for kw, cnt in keywords_cnt:
        if word == kw:
            print(word, cnt)
print('-'*80)
```
